## Cross Correlation in Closure Spectrum Test

Transform two spectra of differing triad class back and perform an eigendecomposition to find orthogonal eigenvalues which correspond to the correlated and uncorrelated parts of the signal.



```python
ps1 = psXmedCTimCTriC(numpy.exp(1j*(pp_eq14_yy[125:225,:,:,120:380])),
                       window="hamming")
ps2 = psXmedCTimCTriC(numpy.exp(1j*(pp_eq28_yy[125:225,:,:,120:380])),
                       window="hamming")
ps3 = psXmedCTimCTriC(numpy.exp(1j*(pp_linew_yy[125:225,:,:,120:380])),
                       window="hamming")

cl1 = numpy.fft.ifft(numpy.fft.fftshift(ps1))
cl2 = numpy.fft.ifft(numpy.fft.fftshift(ps2))
cl3 = numpy.fft.ifft(numpy.fft.fftshift(ps3))



cov1 = numpy.cov(cl1,cl2)
cov2 = numpy.cov(cl2,cl3)
eigval,eigvec = numpy.linalg.eig(cov1)
eigval2,eigvec2 = numpy.linalg.eig(cov2)
stack = numpy.vstack((cl1,cl2)).T
stack2 = numpy.vstack((cl2,cl3)).T
sumeigs = numpy.dot(stack,eigvec)
sumeigs2 = numpy.dot(stack2,eigvec2)



pss1 = numpy.fft.fft(sumeigs[:,0])
pss2 = numpy.fft.fft(sumeigs[:,1])
pss3 = numpy.fft.fft(sumeigs2[:,0])
pss4 = numpy.fft.fft(sumeigs2[:,1])

#pp_ps(numpy.fft.fftshift(numpy.abs(pss1)),title='Correlated Noise, EQ14/EQ28')
#pp_ps(numpy.fft.fftshift(numpy.abs(pss2)),title='Uncorrelated Noise, EQ14/EQ28')

#pp_ps(numpy.fft.fftshift(numpy.abs(pss3)),title="Correlated Noise, EQ28/LinEW")
#pp_ps(numpy.fft.fftshift(numpy.abs(pss4)),title="Uncorrelated Noise, EQ28/LinEW")

fig=mkfig()

ax=fig.add_subplot(111)
xvar = numpy.arange(128)-64

pylab.semilogy(xvar,numpy.fft.fftshift(numpy.abs(pss1)),label = 'EQ14/EQ28 YY, Correlated')
pylab.semilogy(xvar,numpy.fft.fftshift(numpy.abs(pss2)),label = 'EQ14/EQ28 YY, Uncorrelated')
pylab.semilogy(xvar,numpy.fft.fftshift(numpy.abs(pss3)),label = 'EQ28/LinEW YY, Correlated')
pylab.semilogy(xvar,numpy.fft.fftshift(numpy.abs(pss4)),label = 'EQ28/LinEW YY, Uncorrelated')

ax.set_xlabel("Delay ( %f ns)" % (1e9/(100e6/1024) / 128,))
ax.set_ylabel("$\lambda$")
ax.set_title("XX")
ax.set_ylim([10e-9,10e0])
ax.legend(loc=2)
ax.xaxis.set_ticks_position('bottom')
ax.grid()

```
